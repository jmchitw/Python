***Code below is testing the conversion process of converting the string values into an integer totaling the number of 100th seconds. In this example, I was able to use the split function to separate the minute from the seconds and passing that to another split that separated seconds and hundredths of seconds. What I discovered was the function is able to recognize the previous functions output as an indexed list.***

```python
temp = times[0]
[min, secs, hundredths] = [temp.split(":")[0], temp.split(":")[1].split(".")[0], 
                           temp.split(":")[1].split(".")[1]]
converted_time = (((int(min) * 60) + int(secs)) * 100) + int(hundredths)

converted_time
```

***The same conversion using the method illustrated in the book.***

```python
first = times[0]
minutes, rest = first.split(":")
seconds, hundredths = rest.split(".")
converted_time = (int(minutes) * 60 * 100) + (int(seconds) * 100) + int(hundredths)

converted_time
```

***Code to create a list of all converted times to hundredths*** 

```python
converted_times = []
for t in times:
    [min, secs, hundredths] = [t.split(":")[0], 
                               t.split(":")[1].split(".")[0], 
                               t.split(":")[1].split(".")[1]]
    converted_times.append((((int(min) * 60) + int(secs)) * 100) + int(hundredths))

converted_times 

```

***Using the Python statistics module to compute the average swim times***

```python
import statistics
statistics.mean(converted_times)
```

```python
print(dir(statistics))
```

***Converting average swim times to time string (minutes:seconds.hundredths)***

```python
total_seconds, hundredths = str(round((statistics.mean(converted_times)) / 100, 2)).split(".")
minutes = int(total_seconds) // 60
seconds = int(total_seconds) % 60
average_time = str(minutes) + ":" + str(seconds) + "." + hundredths
average_time
```
