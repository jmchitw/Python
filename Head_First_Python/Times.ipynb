```python
import statistics

FN = "Darius-13-100m-Fly.txt"
FOLDER = "swimdata/"

[swimmer, age, distance, stroke] = FN.removesuffix(".txt").split("-")

with open (FOLDER + FN) as file:
    lines = file.readlines()
    times = lines[0].strip().split(",")
  
converted_times = []
for t in times:
    [min_part, sec_part, hundredths_part] = [t.split(":")[0], 
                               t.split(":")[1].split(".")[0], 
                               t.split(":")[1].split(".")[1]]
    converted_times.append((((int(min_part) * 60) + int(sec_part)) * 100) + int(hundredths_part))

total_seconds, hundredths = str(round((statistics.mean(converted_times)) / 100, 2)).split(".")
minutes = int(total_seconds) // 60
seconds = int(total_seconds) % 60
average_time = str(minutes) + ":" + str(seconds) + "." + hundredths

print(swimmer)
print(age)
print(distance)
print(stroke)
print("TIMES:",times)
print("AVERAGE TIME:", average_time)
```
